```python
import numpy as np
from skimage.io import imread, imshow
import matplotlib.pyplot as plt

import mypackage as mp
```

```python
image = imread("toto.tif").astype(float)
arr = np.zeros((1,image.shape[0],image.shape[1]))
arr[0] = image
print(type(image), image.dtype, image.shape)
plt.imshow(arr[0])
plt.show()
```

```python
cells = mp.rebuild_tissue(arr)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(arr[0])

for c in cells:
    ax.plot(c.centroid[-1], c.centroid[-2], 'ro')  # 'ro' for red dot
    for v in c.vertices:
        ax.plot(v.x, v.y, 'b+')

for c in cells:
    c0 = c.centroid
    for e in c.edges:
        c1 = e.mid_point
        ax.plot([c0[-1], c1[-1]], [c0[-2], c1[-2]], 'r-')
        ax.plot(c1[-1], c1[-2], 'g*')

plt.show()
```
